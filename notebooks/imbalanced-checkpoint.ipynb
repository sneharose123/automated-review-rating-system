## Data Preprocessing

### Create a cleaning function

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_review(text):
    if pd.isna(text):
        return ""

    # make text lowercase
    text = text.lower()

    # remove links and html
    text = re.sub(r'http\S+|www\S+|<.*?>', '', text)

    # remove emojis, numbers, punctuation
    text = re.sub(r'[^a-z\s]', '', text)

    # split words
    words = text.split()

    # remove stopwords and apply lemmatization
    clean_words = []
    for word in words:
        if word not in stop_words:
            clean_words.append(lemmatizer.lemmatize(word))

    return " ".join(clean_words)
```

### Apply Cleaning

```python
imbalanced_df['clean_review'] = imbalanced_df['review_text'].apply(clean_review)
```

### Filter short and long reviews

```python
imbalanced_df['word_count'] = imbalanced_df['clean_review'].apply(lambda x: len(x.split()))
max_words = imbalanced_df['word_count'].max()

print("Maximum word count in a review:", max_words)
```

```python
# keep reviews with 3 to 300 words
imbalanced_df = imbalanced_df[(imbalanced_df['word_count'] >= 3) & (imbalanced_df['word_count'] <= 300)]
```

## Stratified TrainTest Split (8020)

```python
from sklearn.model_selection import train_test_split

X1 = imbalanced_df['clean_review']
y1 = imbalanced_df['rating']

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1,
    y1,
    test_size=0.2,
    random_state=42,
    stratify=y1
)
```

```python
X1_train_clean = X1_train.apply(clean_review)
X1_test_clean = X1_test.apply(clean_review)
```

## Text Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer1 = TfidfVectorizer(
    max_features=15000,
    ngram_range=(1, 2),
    min_df=5,
    max_df=0.8,
    sublinear_tf=True
)
X1_train_vec = vectorizer1.fit_transform(X1_train_clean)
X1_test_vec = vectorizer1.transform(X1_test_clean)
```

```python
print("Train shape:", X1_train_vec.shape)
print("Test shape:", X1_test_vec.shape)
```

## Model Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_lr_imb = LogisticRegression(
    max_iter=5000,
    solver="saga",
    C=2.0,
    class_weight="balanced",
    n_jobs=-1,
    multi_class="multinomial"
)
model_lr_imb.fit(X1_train_vec, y1_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred1 = model_lr_imb.predict(X1_test_vec)

print("Accuracy:", accuracy_score(y1_test, y_pred1))
print("\nClassification Report:\n")
print(classification_report(y1_test, y_pred1))
print("Confusion Matrix:\n", confusion_matrix(y1_test, y_pred1))
```

### Linear SVM

```python
from sklearn.svm import LinearSVC

model_svm_imb = LinearSVC(
    C=1.0,
    class_weight='balanced',
    max_iter=5000,
    random_state=42
)

model_svm_imb.fit(X1_train_vec, y1_train)

y_pred2 = model_svm_imb.predict(X1_test_vec)

print("SVM Accuracy:", accuracy_score(y1_test, y_pred2))
print(classification_report(y1_test, y_pred2))
```

### Ridge classifier

```python
from sklearn.linear_model import RidgeClassifier

model_rc_imb = RidgeClassifier(
    alpha=1.0,
    class_weight='balanced'
)

model_rc_imb.fit(X1_train_vec, y1_train)
```

```python
y_pred3 = model_rc_imb.predict(X1_test_vec)
print("Ridge Accuracy:", accuracy_score(y1_test, y_pred3))
print(classification_report(y1_test, y_pred3))
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report
model_sgd_imb = SGDClassifier(
    loss="log",
    max_iter=1000,
    random_state=42
)

model_sgd_imb.fit(X1_train_vec, y1_train)
```

```python
y_pred4 = model_sgd_imb.predict(X1_test_vec)

print("SGD Accuracy:", accuracy_score(y1_test, y_pred4))
print(classification_report(y1_test, y_pred4))
```

### Logistic Regression Pipeline (Imbalanced)

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
pipe_lr_imb = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression(
        solver="saga",
        max_iter=5000,
        n_jobs=-1,
        multi_class="multinomial",
        class_weight="balanced"
    ))
])
```

### Parameter Grid (Logistic Regression)

```python
param_grid_lr_imb = {
    "tfidf__max_features": [5000, 10000, 15000],
    "tfidf__ngram_range": [(1,1), (1,2)],
    "tfidf__min_df": [3, 5],
    "tfidf__max_df": [0.8, 0.9],
    "clf__C": [0.5, 1.0, 2.0, 3.0]
}
```

## Logistic Regression GridSearchCV

```python
grid_lr_imb = GridSearchCV(
    pipe_lr_imb,
    param_grid_lr_imb,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_lr_imb.fit(X1_train, y1_train)
```

```python
pipe_svm_imb = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LinearSVC(class_weight="balanced"))
])
```

```python
param_grid_svm_imb = {
    "tfidf__max_features": [5000, 10000, 15000],
    "tfidf__ngram_range": [(1,1), (1,2)],
    "clf__C": [0.5, 1.0, 2.0, 3.0]
}
```

## Linear SVM Grid Search

```python
grid_svm_imb = GridSearchCV(
    pipe_svm_imb,
    param_grid_svm_imb,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_svm_imb.fit(X1_train, y1_train)
```

```python
from sklearn.linear_model import SGDClassifier
pipe_sgd = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", SGDClassifier(
        loss="log",
        max_iter=5000,
        random_state=42,
        class_weight="balanced"
    ))
])
```

```python
param_grid_sgd = {
    "tfidf__max_features": [5000, 10000, 15000],
    "tfidf__ngram_range": [(1,1), (1,2)],
    "tfidf__min_df": [3, 5],
    "tfidf__max_df": [0.8, 0.9],
    "clf__alpha": [1e-4, 1e-3, 1e-2],
    "clf__penalty": ["l2", "l1"]
}
```

## SGD Classifier Grid Search

```python
grid_sgd = GridSearchCV(
    pipe_sgd,
    param_grid_sgd,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_sgd.fit(X1_train, y1_train)
```

## Compare Best Models

```python
models = {
    "Logistic Regression": grid_lr_imb,
    "Linear SVM": grid_svm_imb,
    "SGD Classifier": grid_sgd
}

for name, grid in models.items():
    print(f"\n{name}")
    print("Best Macro F1:", grid.best_score_)
    print("Best Params:", grid.best_params_)
```

## Evaluate the Best Model on Test Set

```python
model_B = grid_lr_imb.best_estimator_
print(model_B)
print("Best Parameters:\n", grid_lr_imb.best_params_)

y_pred_imb = model_B.predict(X1_test)

print("Accuracy:", accuracy_score(y1_test, y_pred_imb))
print(classification_report(y1_test, y_pred_imb))
```
